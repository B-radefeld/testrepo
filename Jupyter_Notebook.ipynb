## My code creates a simple parabola

```python
import matplotlib.pyplot as plt
import numpy as np

# create 1000 equally spaced points between -10 and 10
x = np.linspace(-5, 5, 100)

# calculate the y value for each element of the x vector
y = x**2

plt.xlabel('x - axis')
plt.ylabel('y-axis')
plt.title('Parabola')
plt.plot(x, y)
```
